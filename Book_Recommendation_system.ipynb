# POPULARITY BASED (HIGHEST RATING)

merging rating and book df on ISBN

```python
new_df  = rating.merge(book, on = 'ISBN')
```

```python
new_df
```

```python
new_df.shape
```

```python
new_df.isnull().sum()
```

```python
num_rating_df = new_df.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating': 'num_ratings'})
```

```python
avg_rating_df = new_df.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating': 'avg_ratings'})
```

```python
rating_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
rating_df.rename(columns = {'Book-Rating_x': 'num_ratings', 'Book-Rating_y': 'avg_ratings'}, inplace = True)
```

```python
rating_df
```

```python
popular_df = rating_df[rating_df['num_ratings']>=250].sort_values('avg_ratings', ascending = False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(book, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'num_ratings', 'Image-URL-M', 'avg_ratings', 'Book-Author', 'Year-Of-Publication']]
```

```python
popular_df
```

# COLLABORATIVE FILTERING

```python
new_df
```

```python
x = new_df.groupby('User-ID').count()['Book-Rating']> 200
user_with_higher_rating = x[x].index
user_with_higher_rating
# the users who rated books more than 200
```

```python
filtered_rating = new_df[new_df['User-ID'].isin(user_with_higher_rating)]
# users rating with higher than 200
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famouse_books = y[y].index
famouse_books
```

```python
filtered_books = new_df[new_df['Book-Title'].isin(famouse_books)]
```

```python
filtered_books
```

```python
filtered_books.drop_duplicates()
```

```python
pt = filtered_books.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace = True)
```

```python
pt
```

similarity based books

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0] # means the first book 1984's similarity scores with all books and also with itself
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
  index = np.where(pt.index== book_name)[0][0]
  similar_item = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1], reverse = True)[1:6]

  data = []
  for i in similar_item:
    item = []
    print(pt.index[i[0]])
    temp_df = book[book['Book-Title'] == pt.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

    data.append(item)

  return data


# sorted(list(enumerate(similarity_score[0])), key = lambda x:x[1], reverse = True)[1:6]
```

```python
recommend('Harry Potter and the Chamber of Secrets (Book 2)')  # recommended books
```
